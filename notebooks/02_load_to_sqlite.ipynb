```python
# ============================================================
# LOAD CLEANED CSVs → SQLite DATABASE
# ============================================================

import sqlite3
import pandas as pd
import os

# Create database in models folder
db_path = '../data/models/supply_chain.db'
os.makedirs('../data/models', exist_ok=True)
conn = sqlite3.connect(db_path)

print(f"✅ Connected to SQLite at: {db_path}")

# Load each cleaned CSV and write to SQLite
tables = {
    'dim_date':     '../data/cleaned/dim_date.csv',
    'dim_customer': '../data/cleaned/dim_customer.csv',
    'dim_product':  '../data/cleaned/dim_product.csv',
    'dim_geo':      '../data/cleaned/dim_geo.csv',
    'fact_orders':  '../data/cleaned/fact_orders.csv',
}

for table_name, csv_path in tables.items():
    df = pd.read_csv(csv_path)
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    print(f"✅ Loaded {table_name}: {df.shape[0]:,} rows × {df.shape[1]} cols")

# Verify
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(f"\nTables in DB: {cursor.fetchall()}")

conn.close()
print("\n✅ Database ready.")
```
